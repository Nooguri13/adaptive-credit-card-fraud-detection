```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Model F1-score data
data = {
    "Model": [
        "Logistic Regression", "Ridge Classifier", "Passive Aggressive",
        "Decision Tree", "Random Forest", "Extra Trees", "Gradient Boosting",
        "AdaBoost", "Bagging Classifier", "Gaussian Naive Bayes", "Linear SVC",
        "K-Nearest Neighbors", "KMeans Clustering (k=2)", "Gaussian Mixture Model",
        "Isolation Forest", "Local Outlier Factor", "Elliptic Envelope",
        "XGBoost", "Neural Network (MLP)", "Q-Table", "Deep Q-Learning (DQN)"
    ],
    "F1-Score": [
        0.72, 0.68, 0.64, 0.76, 0.89, 0.91, 0.87, 0.82, 0.84, 0.58,
        0.72, 0.75, 0.45, 0.49, 0.68, 0.71, 0.56, 0.92, 0.87, 0.51, 0.97
    ]
}

df = pd.DataFrame(data)

# Identify best model
best_model = df.loc[df["F1-Score"].idxmax()]
best_f1 = best_model["F1-Score"]
best_label = f"Best Model: {best_model['Model']} (F1={best_f1:.2f})"

# Sort for better plot
df = df.sort_values(by="F1-Score", ascending=False)

# Plotting
plt.figure(figsize=(14, 7))
sns.barplot(x="F1-Score", y="Model", data=df, palette="Blues_r")
plt.axvline(best_f1, color='red', linestyle='--', label=best_label)
plt.title("Best Model Based on F1-Score", fontsize=16)
plt.xlabel("F1-Score")
plt.ylabel("Model")
plt.legend()
plt.tight_layout()
plt.show()
```
